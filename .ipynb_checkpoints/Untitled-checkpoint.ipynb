```python
class MNISTDataset(data.Dataset):
    def __init__(self, X:np.ndarray, y:np.ndarray, transform=None):
        self.X = X.reshape(-1, 28, 28, 1).astype(np.float32).copy()
        self.y = y.astype(np.int64).copy()
        self.transform = transform
        
    def __getitem__(self, index):
        x = self.X[index].copy()
        
        if self.transform is not None:
            x = self.transform(x)
        y = self.y[index]
        y = torch.Tensor([y]).long().squeeze()
        # print(x.shape, y.shape)
        return(x,y)
    
    def __len__(self):
        return(self.X.shape[0])         
```

```python
def conv3x3(in_channels,out_channels):
    return(nn.Conv2d(in_channels, out_channels, 3, stride=1, padding=1))

def conv_relu(in_channels, out_channels):
    return(nn.Sequential(conv3x3(in_channels, out_channels), nn.ReLU()))

class MNISTCnn(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.conv1 = conv_relu(1,8)
        self.conv2 = conv_relu(8,64)
        self.conv3 = conv_relu(64,256)
        self.avgpool = nn.AdaptiveAvgPool2d(1)
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(256, 10)
        self.maxpool = nn.MaxPool2d(2, stride=2)
        
    def forward(self, x):
        x = self.conv1(x)
        x = self.maxpool(x)
        x = self.conv2(x)
        x = self.maxpool(x)
        x = self.conv3(x)
        x = self.avgpool(x)
        x = x.squeeze()
        x = self.dropout(x)
        return(self.fc(x))
```

```python
model = MNISTCnn()
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print('Model parameters: {:,}'.format(num_params))
```

```python
xform = transforms.Compose([transforms.ToTensor(),
                           transforms.Normalize([x_mean], [x_std])])
datasets = {}

datasets['train'] = MNISTDataset(X_train, y_train, transform=xform)
datasets['val'] = MNISTDataset(X_test, y_test, transform=xform)
loaders = {}
loaders['train'] = data.DataLoader(datasets['train'], batch_size=64, shuffle=True, num_workers=0, pin_memory=True)
loaders['val'] = data.DataLoader(datasets['val'], batch_size=64, shuffle=True, num_workers=0, pin_memory=True)
```

```python
device = torch.device('cuda:0')
# torch.set_device(device)
```

```python
lr = 1e-4
optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=lr)
```

```python
train_losses = []
val_losses = []
accs = []
model = model.to(device)


for epoch in tqdm(range(100)):
    model.train()
    dataiter = iter(loaders['train'])
    
    t = tqdm(dataiter, leave=False)
    epoch_losses = []
    for images, labels in t:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        loss_int = loss.item()
        epoch_losses.append(loss_int)
        t.set_description('loss: {:.4f}'.format(loss_int) )
        
    train_losses.append(np.array(epoch_losses).mean())
    
    model.eval()
    dataiter = iter(loaders['val'])
    t = tqdm(dataiter, leave=False)
    N=0
    correct = 0
    epoch_losses = []
    for images, labels in t:
        images,labels = images.to(device), labels.to(device)
        with torch.no_grad():
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            predictions = torch.softmax(outputs, dim=1)
            hardmax = torch.argmax(predictions, dim=1)
            equals = (hardmax==labels).to(int)
            correct += equals.sum().item()
            N += len(equals)
            running_acc = correct / N
            
            epoch_losses.append(loss.item())
        t.set_description('val acc: {:.4f}'.format(running_acc) )
    accs.append(correct/N)
    val_losses.append(np.array(epoch_losses).mean())
    
```

```python
x = np.arange(len(val_losses))
y = np.array(val_losses)

fig, ax = plt.subplots(1,1)
ax.plot(x,y)
ax.set_ylabel('val loss')
plt.show()
```

```python
def make_columnar_image(a, height=224):
    if type(a) == torch.Tensor:
        a = a.detach().cpu().numpy()
    shape = a.shape
    # assert(len(shape)==3)
    if len(shape) == 3:
        n, h, w = shape
        n_columns = np.ceil(n*h/height).astype(int)
        inds = np.linspace(0, n, n_columns+1).astype(int)
        starts = inds[:-1]
        ends = inds[1:]
        columns = []
        for start, end in zip(starts,ends):
            columns.append(np.vstack((a[start:end,...])))
        im = np.hstack((columns))
    elif len(shape) < 3:
        im = np.vstack((a))
    # print(im.shape)
    im = im.squeeze()
    if im.ndim < 2:
        im = im[:, np.newaxis]
    if im.shape[0] < height:
        aspect = im.shape[1]/im.shape[0]
        aspect = max(aspect, 0.1)
        new_height = height
        new_width = int(aspect*new_height)
        im = cv2.resize(im, (new_width, new_height), interpolation=cv2.INTER_NEAREST)
    return(im)
```

```python
def join_columns(imlist, pad=10, cmap='gray'):
    cmap = plt.get_cmap(cmap)
    N = len(imlist)
    H = imlist[0].shape[0]
    new_list = []
    im = cmap(imlist[0])
    zeros = np.zeros_like(cmap(imlist[0]), shape=(H,pad))
    for i in range(N):
        im = imlist[i]
        im -= im.min()
        im /= im.max()
        new_list.append(im)
        # new_list.append(cmap(im))
        new_list.append(zeros)
    new_list = new_list[:-1]
    return(np.hstack((new_list)))
```

```python
def visualize_batch(activations, images, names, N_in_batch):
    subdict = {k:v[N_in_batch] for k, v in activations.items()}
    ims = [make_columnar_image(subdict[name]) for name in names]
    activation_image = join_columns(ims)
    height = activation_image.shape[0]
    
    input_image = images[N_in_batch,...].detach().cpu().numpy().squeeze()
    input_image -= input_image.min()
    input_image /= input_image.max()
    padval = int((224-28)/2)
    input_padded = np.pad(input_image, ((padval, padval), (0,0)))

    zeros = np.zeros_like(input_padded, shape=(224,10))
    output_image = np.hstack((input_padded, zeros, activation_image))
    return(output_image)
```
